## Python Functions

```python
def show_df(filename='olympics_1896_2004.csv'):
  oo = pd.read_csv(filename, engine='python', skiprows=5, on_bad_lines='skip')
  return oo.head()

show_df()
```

```python
type(show_df())
```

```python
def preprocess():
  ordered_medals = pd.api.types.CategoricalDtype(categories=['Bronze', 'Silver', 'Gold'], ordered=True)
  dtype_mapper = {
    'Year': 'int64',
    'City': 'string',
    'Sport': 'string',
    'Discipline': 'string',
    'Athlete Name': 'string',
    'NOC': 'string',
    'Country': 'string',
    'Gender': 'category',
    'Event Gender': 'category',
    'Medal': ordered_medals
  }
  df = pd.read_csv('olympics_1896_2004.csv', engine='python', skiprows=5, on_bad_lines='skip', dtype=dtype_mapper).drop('Position', axis=1)
  df['Event Gender'] = df['Event Gender'].astype('category')
  return df

oo = preprocess()
oo.head()
```

## Working with Dates

```python
start_date = ['07-27-2012', '5th Aug 2016', '23rd jul 2021']
end_date = ['12th Aug 2012', '21-08-2016', '8th Aug 2021']
```

```python
games = pd.DataFrame(zip(city, start_date, end_date), columns=['city', 'start_date', 'end_date'])
```

## Combining Dataframes

```python
start = pd.DataFrame({
    "city": city,
    "start_date": start_date
})

end = pd.DataFrame({
    "city": city,
    "end_date": end_date
})
```

```python
start
```

```python
end
```

```python
pd.concat([start, end], axis=0)
```

```python
pd.concat([start, end], axis=1)
```

```python
start.loc[3] = ['Paris', '24-8-2024']
end.loc[3] = ['Beijing', '19-07-2008']
```

```python
pd.merge(start, end, on='city')
```
